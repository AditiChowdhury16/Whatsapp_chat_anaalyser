```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with BMIS (Project Group).txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
# pattern = r'(\d{2}/\d{2}/\d{4}),\s(\d{2}:\d{2})'
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
df['message_date'] = pd.to_datetime(df['message_date'].str.extract(r'(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2})')[0], format='%d/%m/%Y, %H:%M')


df.rename(columns={'message_date':'date'},inplace=True)
df.head()
```

```python
df.shape
```

```python
#seperate user name and message
users=[]
messages=[]
for message in df['user_message']:
   entry = re.split('([\w\W]+?):\s', message)
   if entry[1:]:     # user name
     users.append(entry[1])
     messages.append(entry[2])
   else:
     users.append('group_notification')
     messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
words=[]
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
          words.append(word)
    
```
